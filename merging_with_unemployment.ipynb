```python
import pandas as pd
```

```python
suicide_df = pd.read_csv("data/suicide_cleaned.csv")
unemployment_df = pd.read_csv("data/unemployment.csv")
unemployment_df.head()
```

```python
suicide = suicide_df.drop(columns=['population', 'suicide'])
suicide.head()
```

```python
unemployment = unemployment_df[unemployment_df['Country Name'].isin(suicide['country'].unique()) ].drop(columns='Country Code').rename(columns={'Country Name':'country'})
unemployment = unemployment.set_index('country').melt(ignore_index=False).rename(columns={'variable':'year', 'value':'unemployment_rate'})
unemployment['year'] = unemployment['year'].astype(int)
unemployment = unemployment[ unemployment['year'] <= 2015 ].reset_index()
unemployment.head()
```

```python
output_df = pd.merge(suicide, unemployment, on=['country', 'year'])
output_df.head()
```

```python
output_df.to_csv('data/suicide_with_unemploy.csv')
```
